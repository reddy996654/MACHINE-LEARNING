```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import numpy as np

# 1. Load Dataset
digits = load_digits()
X = digits.data
y = digits.target

# 2. Implement PCA
# Reduce 64 dimensions to 2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3. Display Numerical Output
print(f"Original shape: {X.shape}")
print(f"Reduced shape: {X_pca.shape}")
print("\nExplained Variance Ratio by 2 components:")
print(pca.explained_variance_ratio_)
print(f"\nTotal variance explained: {np.sum(pca.explained_variance_ratio_) * 100:.2f}%")

# 4. Create Visualization (Graph)
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Paired', alpha=0.8)
plt.title('PCA of Digits Dataset (64D -> 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(handles=scatter.legend_elements()[0], labels=digits.target_names, title="Digits")
plt.grid(True)
plt.show()
```
