```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import pandas as pd

# 1. Load Dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# We'll just use the first two features for easy 2D visualization
X_plot = X[:, :2]

# 2. Implement Algorithms
# K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_plot)

# Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X_plot)

# Hierarchical Clustering (Agglomerative)
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X_plot)

# 3. Create Visualizations (Graphs)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Clustering Algorithm Comparison on Iris Dataset', fontsize=16)

# Plot 1: Ground Truth
axes[0, 0].scatter(X_plot[:, 0], X_plot[:, 1], c=y, cmap='viridis', s=50)
axes[0, 0].set_title('Ground Truth (Actual Species)')
axes[0, 0].set_xlabel(feature_names[0])
axes[0, 0].set_ylabel(feature_names[1])

# Plot 2: K-Means
axes[0, 1].scatter(X_plot[:, 0], X_plot[:, 1], c=kmeans_labels, cmap='viridis', s=50)
axes[0, 1].set_title('K-Means Clustering')
axes[0, 1].set_xlabel(feature_names[0])
axes[0, 1].set_ylabel(feature_names[1])

# Plot 3: Gaussian Mixture Model (GMM)
axes[1, 0].scatter(X_plot[:, 0], X_plot[:, 1], c=gmm_labels, cmap='viridis', s=50)
axes[1, 0].set_title('Gaussian Mixture Model (GMM)')
axes[1, 0].set_xlabel(feature_names[0])
axes[1, 0].set_ylabel(feature_names[1])

# Plot 4: Hierarchical Clustering
axes[1, 1].scatter(X_plot[:, 0], X_plot[:, 1], c=hierarchical_labels, cmap='viridis', s=50)
axes[1, 1].set_title('Hierarchical Clustering')
axes[1, 1].set_xlabel(feature_names[0])
axes[1, 1].set_ylabel(feature_names[1])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
