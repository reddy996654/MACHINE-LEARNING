```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# 1. Load Dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
feature_names = cancer.feature_names
class_names = cancer.target_names

# 2. Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 3. Implement CART (DecisionTreeClassifier)
# 'criterion="gini"' is the default for CART
cart_model = DecisionTreeClassifier(random_state=42, max_depth=3) # Limit depth for readability
cart_model.fit(X_train, y_train)

# 4. Make Predictions and Get Numerical Output
y_pred = cart_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"--- CART Model ---")
print(f"Accuracy on test set: {accuracy * 100:.2f}%")

# 5. Create Visualization (Graph of the Tree)
plt.figure(figsize=(20, 10))
plot_tree(cart_model,
          filled=True,
          feature_names=feature_names,
          class_names=class_names,
          rounded=True,
          fontsize=10)
plt.title("CART Decision Tree for Breast Cancer Categorization (max_depth=3)")
plt.show()
```
