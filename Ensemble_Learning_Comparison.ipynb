```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# 1. Load and Split Data
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 2. Implement Models

# Model 1: Single Decision Tree (from Exp 8)
tree_model = DecisionTreeClassifier(random_state=42, max_depth=3)
tree_model.fit(X_train, y_train)
tree_acc = accuracy_score(y_test, tree_model.predict(X_test))

# Model 2: Random Forest (Ensemble)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf_model.predict(X_test))

# Model 3: Gradient Boosting (Ensemble)
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
gb_acc = accuracy_score(y_test, gb_model.predict(X_test))

# 3. Display Numerical Outputs
print("--- Model Accuracy Comparison ---")
print(f"Single Decision Tree: {tree_acc * 100:.2f}%")
print(f"Random Forest: {rf_acc * 100:.2f}%")
print(f"Gradient Boosting: {gb_acc * 100:.2f}%")

# 4. Create Visualization (Bar Graph)
model_names = ['Decision Tree', 'Random Forest', 'Gradient Boosting']
accuracies = [tree_acc, rf_acc, gb_acc]

plt.figure(figsize=(10, 6))
bars = plt.bar(model_names, accuracies, color=['orange', 'green', 'blue'])
plt.title('Ensemble Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0.9, 1.0) # Zoom in on the top accuracies

# Add accuracy numbers on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.005, f'{yval*100:.2f}%', ha='center', va='bottom')

plt.show()
```
